Задача 4.1

Дано: $A_1 x + B_1 y+ C_1 z + D_1 = 0$

Для того чтобы записать уравнение плоскости параллельной данной необходимо и достаточно чтобы система уравнений вида:
$\begin{cases}
    A_1 x + B_1 y+ C_1 z + D_1 = 0\\
    A_2 x + B_2 y+ C_2 z + D_2 = 0
\end{cases}$

не имела решений.
Если такая система не имеет решений,то не существует ни одной точки в прямоугольной системе координат $O_{xyz}$ в трехмерном пространстве, координаты которой удовлетворяют одновременно обоим уравнениям системы.
Рассмотрим две матрицы составленные из системы:

$M = 
 \begin{pmatrix}
  A_1 & B_1 & C_1 \\
  A_2 & B_2 & C_2 \\ 
 \end{pmatrix}$
и расширенную:
$\widetilde{M} = 
 \begin{pmatrix}
  A_1 & B_1 & C_1 & D_1\\
  A_2 & B_2 & C_2 & D_2\\ 
 \end{pmatrix}$
Если все определители основной матрицы будут равны нулю и следовательно ранг основной матрицы будет равен 1. А определитель расширенной матрицы, к примеру:
$\begin{vmatrix}
  A_1 & D_1\\
  A_2 & D_2\\ 
 \end{vmatrix} = A_1 D_2 - A_2 D_1 \neq 0\\
 A_1 D_2 \neq A_2 D_1$
следовательно ранг расширенной равен двум. Если ранг основной матрицы системы уравнений меньше ранга расширенной матрицы системы, то согласно теореме Кронекера-Капелли следует, что система уравнений решений не имеет, тогда плоскости будут параллельны.

Задача 4.2

$A_1 x + B_1 y+ C_1 z + D_1 = 0$

$L : \frac{x-x_1}{x_2 - x_1} = \frac{y-y_1}{y_2 - y_1} = \frac{z-z_1}{z_2 - z_1}$

чтобы доказать что прямая принадледит плоскости необходимо доказать что две точки прямой принадлежат плоскости, иначе говоря должны выполняться следующие равенства:
$\begin{cases}
    A_1 x_1 + B_1 y_1+ C_1 z_1 + D_1 = 0\\
    A_2 x_1 + B_2 y_2+ C_2 z_2 + D_2 = 0
\end{cases}$
, где $(x_1 ,y_1 , z_1), (x_2 ,y_2 , z_2) \in L$

```python
print('Задание 4')
display(Math(r"""
\begin{cases}
    y = x^2 - 1\\
    e^x +x(1-y) = 1
\end{cases}"""))
x=np.linspace(-2,3,201)
plt.plot(x,(np.exp(x)-1)/x +1)
plt.plot(x, x**2 -1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,7)
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y-x**2 +1, np.exp(x) +x - x*y -1)

x1,y1 = fsolve(equations, (-2,1))
print(x1,y1)

x2,y2 = fsolve(equations, (2,2))
print(x2,y2)
```
