```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math, Latex
%matplotlib inline

print('Задание 1')
A = (1,2)
B = (3,10)
C = (5,1)
print('A', A,'; B', B,'; C', C)
a = (A[0] - (C[0]*(B[1]-A[1])+B[0]*A[1] - A[0]*B[1])/(B[0]-A[0]))/(C[0]*(C[0]-A[0]-B[0])+A[0]*B[0])
display(Math(r"""a = \frac{y_C - \frac{x_C (y_B - y_A) + x_B y_A - x_A y_B}{x_B - x_A}}{x_C (x_C - x_A - x_B) + x_A x_B} \\
  = \frac{1 - \frac{5 (10 - 2) + 3*2 - 1*10}{3 - 1}}{5 (5 - 1 - 3) + 1*3} \\
  = """ + str(a)))
b = (B[1]-A[1])/(B[0]-A[0]) - a *(A[0]+B[0])
display(Math(r"""b = \frac{y_B - y_A}{x_B - x_A} - a(x_A + x_B) \\
  = \frac{10 - 2}{3 - 1} - a(1 + 3) \\
  = """ + str(b)))
c = (B[0]*A[1]-A[0]*B[1])/(B[0]-A[0]) + a *A[0]*B[0]
display(Math(r"""c = \frac{x_B y_A - x_A y_B}{x_B - x_A} - a x_A x_B) \\
  = \frac{3*2 - 10*1}{3 - 1} - a*1*3 \\
  = """ + str(c)))
print(f"""\nУравнение прямой:{a} x^2 + ({b}) x + ({c}) = 0""")
xarr = [i for i in range(-50, 50)]
data = [(a*x**2 + b*x +c) for x in xarr]


        
# mu, sigma = 1.1, 2.9
# data = np.random.normal(mu, sigma, 10000)

fig, ax = plt.subplots()

ax.plot(data)

# ax.set_title(r'$p(x)=\frac{1}{\sqrt{2\pi\sigma^{2}}}e^{-\frac{(x-\mu)^{2}}{2\sigma^{2}}}$')
ax.set_title(f"""{a} x^2 + ({b}) x + ({c}) = 0""")


plt.show()
```
